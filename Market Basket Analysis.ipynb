```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_excel('productdata.xls', sheet_name='data')
df.head()
```

```python
df = df.iloc[ : , 4:]
df.head(1)
```

```python
df = df.applymap(lambda x: True if x=="yes" else False)
df.head(2)
```

```python
frequent_datasets = apriori(df, min_support=0.2, use_colnames=True)
frequent_datasets
```

```python
rules = association_rules(frequent_datasets, metric='confidence', min_threshold=0.4)
rules[0:5]
```

```python
rules_1 = association_rules(frequent_datasets, metric='lift', min_threshold=1)
rules_1[20:25]
```
